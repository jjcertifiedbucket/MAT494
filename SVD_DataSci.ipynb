```python
#Calling libraries
import numpy as np
import scipy.linalg as la
from matplotlib.image import imread
import matplotlib.pyplot as plt


#Importing an image
A = imread('/Users/Justin/Desktop/FALL2022/MAT494/JupyterNotebook/puppy Small.png')
X = np.mean(A, axis = -1)
X_noise = X + np.random.normal(size = X.shape)
U, S, VT = la.svd(X)
S = np.diag(S)
X_appx = U[:,:25]@S[0:25,:25]@VT[:25,:]
U1, S1, VT1 = la.svd(X_noise)
S1 = np.diag(S1)
X_appx1 = U1[:,:25]@S1[0:25,:25]@VT1[:25,:]
```

```python
#Plotting
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
img = plt.imshow(X)
plt.title('Original')
img.set_cmap('gray')
plt.subplot(2,2,2)
img = plt.imshow(X_noise)
img.set_cmap('gray')
plt.title('Noisy Image')
plt.subplot(2,2,3)
img = plt.imshow(X_appx)
img.set_cmap('gray')
plt.title('25-rank approximated image')
plt.subplot(2,2,4)
img = plt.imshow(X_appx1)
img.set_cmap('gray')
```
