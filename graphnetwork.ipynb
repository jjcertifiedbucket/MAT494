```python
import networkx as nx
```

```python
# default is 3x3 = 9 nodes
G = nx.sudoku_graph()
print(G.number_of_nodes())
print(G.number_of_edges())
print("--------------------")
A = nx.adjacency_matrix(G)
print(A.todense())
nx.draw(G)
```

```python
# Creating 2x2
G = nx.sudoku_graph(2)
G.number_of_nodes()
16
G.number_of_edges()
print("--------------------")
A = nx.adjacency_matrix(G)
print(A.todense())
nx.draw(G)
```

```python
G = nx.grid_graph(dim=(2, 3, 4))
nx.draw(G)
A = nx.adjacency_matrix(G)
print(A)
```
