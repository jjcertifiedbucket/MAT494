```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
def exampleFunc(x):
  return 0.001*np.power(x,4) + 0.2*np.power(x,3) + 1.4*np.power(x,2) + 300*np.sin(x*0.3) + 400*np.sin(0.05*np.power(x,2) + 0.3*x + 2) + 3

x = np.random.uniform(-15,15,15) # 15 Uniformly random numbers between -15 and +15
# x = np.random.uniform(-15,15,10000) # 100 Uniformly random numbers between -15 and +15
yraw = exampleFunc(x)

xtrue = np.linspace(-15, 15, 1000)
ytrue = exampleFunc(xtrue)
plt.figure(1)
plt.plot(xtrue, ytrue, '-k', linewidth=2) # The underlying function itself
plt.scatter(x, yraw)
plt.show()
```

```python
y = yraw + np.random.normal(size=yraw.shape)*150
plt.figure(2)
plt.plot(xtrue, ytrue, '-k', linewidth=2) # The underlying function itself
plt.scatter(x, y)
plt.show()
```

```python
model = keras.Sequential()
model.add(keras.Input(shape=(1,)))
model.add(layers.Dense(1000, activation="elu")) 
model.add(layers.Dense(1, activation="linear"))
model.summary()
```
